```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
messy = pd.read_csv("files/loan.csv")
cleaned = pd.read_csv("files/cleaned_loan.csv")
cleaned.drop("Loan_ID",axis=1,inplace=True)
cleaned.head()
```

null

```python
sns.heatmap(messy.isnull())
plt.show()
```

after

```python
sns.heatmap(cleaned.isnull())
plt.show()
```

outlayer

```python
plt.subplot(4,1,1)
sns.boxplot(x=messy["ApplicantIncome"],data=messy,color="r",label="ApplicantIncome")

plt.subplot(4,1,2)
sns.boxplot(x=messy["CoapplicantIncome"],data=messy,color ="g",label="CoapplicantIncome") 

plt.subplot(4,1,3)
sns.boxplot(x=messy["LoanAmount"],data=messy,color="b",label="LoanAmount")

plt.subplot(4,1,3)
sns.boxplot(x=messy["Loan_Amount_Term"],data=messy,color="b",label="Loan_Amount_Term")


plt.tight_layout()
plt.show()
```

after

```python


plt.subplot(3,1,1)
sns.boxplot(x=cleaned["ApplicantIncome"],data=cleaned,color="r",label="ApplicantIncome")

plt.subplot(3,1,2)
sns.boxplot(x=cleaned["CoapplicantIncome"],data=cleaned,color ="g",label="CoapplicantIncome") 

plt.subplot(3,1,3)
sns.boxplot(x=cleaned["LoanAmount"],data=cleaned,color="b",label="LoanAmount")

plt.legend()
plt.tight_layout()
plt.show()
```

understanding and identifying 

```python
cleaned.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(cleaned['ApplicantIncome'], kde=True)
plt.title('Applicant Income Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Loan_Status', data=cleaned)
plt.title('Loan Status Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Education', hue='Loan_Status', data=cleaned)
plt.title('Education vs Loan Approval')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Loan_Status', data=cleaned)
plt.title('Gender vs Loan Approval')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Married', hue='Loan_Status', data=cleaned)
plt.title('Marital Status vs Loan Approval')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cleaned.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
